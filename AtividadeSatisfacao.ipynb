```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('pesquisa_satisfacao.csv',  sep = ';')
df
```

```python
dfTempo= df.sort_values('Tempo de Espera (min)')
#dfTempo
```

```python
dados = np.array(dfTempo['Tempo de Espera (min)'])
dados
```

```python
q1 = np.percentile(dados, 25)
q2 = np.percentile(dados, 50)
q3 = np.percentile(dados, 75)
```

```python
print(f'Primeiro quartil (Q1): {q1}')
print(f'Segundo quartil (Q2): {q2}')
print(f'Terceiro quartil (Q3): {q3}')
```

```python
#Clientes que estão entre os 25% que receberam mais rápido

df_rapido = dfTempo.loc[(dfTempo['Tempo de Espera (min)'] <= q1)]
df_rapido
```

```python
df_rapido.plot(kind = 'bar' , x= 'Cliente', y= 'Tempo de Espera (min)') 
plt.xlabel('Cliente') 
plt.ylabel('Tempo de Espera (min)') 
plt.title('Clientes entre os 25% que receberam mais rápido') 
plt.show()
```
