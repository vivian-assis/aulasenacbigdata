```python
dados = {
'Departamento': ['Vendas', 'Vendas', 'RH', 'RH', 'TI', 'TI'],
'Funcionário': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eduardo',
'Fernanda'],
'Salário': [3000, 3200, 2500, 2700, 4000, 4200]
}
df = pd.DataFrame(dados)

df
```

```python
df.groupby('Departamento').sum(numeric_only=True)

```

```python
df.groupby('Departamento').mean(numeric_only=True) #agrupamento
```

```python
df_departamento = df.groupby('Departamento')['Salário'].mean() #agrupamento selecionando qual coluna para fazer a média nesse caso
df_departamento
```

```python
df_departamento = df.groupby('Departamento')['Salário'].mean().reset_index()  #agrupamento selecionando qual coluna para fazer a média nesse caso
df_departamento
```
