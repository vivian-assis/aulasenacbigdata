```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv',  sep = ';', encoding= 'latin1')
df.tail()

```

```python
df.info()
```

```python
df_Regioes = df.groupby(['ano','munic'])['furto_celular'].sum().reset_index()
df_Regioes
```

```python
regiao_metropolitana = ['Rio de Janeiro', 'Duque de Caxias','Nova Iguaçu','São Gonçalo','Niterói','Belford Roxo','São João de Meriti','Magé','Itaboraí','Mesquita','Nilópolis','Queimados','Seropédica','Japeri','Paracambi','Guapimirim','Maricá','Tanguá','Itaguaí','Cachoeiras de Macacu','Rio Bonito','Paty do Alferes']
df_Metropole2024 = df_Regioes.loc[(df_Regioes['munic'].isin(regiao_metropolitana))& (df_Regioes['ano']==2024)]
df_Metropole2024 = df_Metropole2024.sort_values(by='furto_celular',ascending=False)
df_Metropole2024
```

```python
array_furtocel = np.array(df_Metropole2024['furto_celular'])
```

```python
media_furtocel = np.mean(array_furtocel)
media_furtocel
```

```python
mediana_furtocel = np.median(array_furtocel)
mediana_furtocel
```

```python
distancia = (media_furtocel - mediana_furtocel)/mediana_furtocel
distancia
```
